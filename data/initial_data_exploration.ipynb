```python
import rasterio
import numpy as np
from rasterio.enums import Resampling
import os
import matplotlib.pyplot as plt

def average_pooling(data, pool_size):
    """Perform average pooling on the input data."""
    output_shape = (data.shape[0] // pool_size, data.shape[1] // pool_size)
    pooled_data = data.reshape(output_shape[0], pool_size, 
                               output_shape[1], pool_size).mean(axis=(1, 3))
    return pooled_data

def process_tif_file(input_path, output_path, pool_size):
    with rasterio.open(input_path) as src:
        data = src.read(1)
        nodata = src.nodata

        data = np.where(data == nodata, np.nan, data)

        pooled_data = average_pooling(data, pool_size)
        pooled_data = np.where(np.isnan(pooled_data), nodata, pooled_data)
        pooled_data = pooled_data.astype(np.int16)
        
        
        transform = src.transform * src.transform.scale(
            (src.width / pooled_data.shape[1]),
            (src.height / pooled_data.shape[0])
        )
        
        new_meta = src.meta.copy()
        new_meta.update({
            'height': pooled_data.shape[0],
            'width': pooled_data.shape[1],
            'transform': transform,
            'dtype': 'int16',
            'nodata': nodata
        })
        
        with rasterio.open(output_path, 'w', **new_meta) as dst:
            dst.write(pooled_data, 1)
    
    return data.shape, pooled_data.shape

# Set the input and output directories
input_dir = "sedac-popdensity-yeargrid5yr-v4.11"
output_dir = "sedac-popdensity-yeargrid5yr-v4.11/pooled"

# Create the output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Set the pooling size
pool_size = 32

# List all files in the input directory
all_files = os.listdir(input_dir)
print(f"Files in input directory: {all_files}")

# Filter for .tif files
tif_files = [f for f in all_files if f.endswith(".tif")]
print(f"TIF files found: {tif_files}")

if not tif_files:
    print("No .tif files found in the input directory. Please check the directory path and file extensions.")
else:
    # Process all TIF files in the input directory
    for filename in tif_files:
        input_path = os.path.join(input_dir, filename)
        output_path = os.path.join(output_dir, f"pooled_{filename}")
        
        original_shape, pooled_shape = process_tif_file(input_path, output_path, pool_size)
        
        print(f"Processed: {filename}")
        print(f"  Original shape: {original_shape}")
        print(f"  Pooled shape: {pooled_shape}")
        print(f"  Size reduction: {original_shape[0] * original_shape[1] / (pooled_shape[0] * pooled_shape[1]):.2f}x")
        print()

    # Visualize the first TIF file before and after pooling
    first_tif = tif_files[0]
    input_path = os.path.join(input_dir, first_tif)
    output_path = os.path.join(output_dir, f"pooled_{first_tif}")

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 7))

    with rasterio.open(input_path) as src:
        ax1.imshow(src.read(1), cmap='viridis')
        ax1.set_title('Original Data')

    with rasterio.open(output_path) as src:
        ax2.imshow(src.read(1), cmap='viridis')
        ax2.set_title('Pooled Data')

    plt.show()

    # Print additional information about the first file
    with rasterio.open(input_path) as src:
        print(f"Original file: {first_tif}")
        print(f"CRS: {src.crs}")
        print(f"Bounds: {src.bounds}")

    with rasterio.open(output_path) as src:
        print(f"\nPooled file: pooled_{first_tif}")
        print(f"CRS: {src.crs}")
        print(f"Bounds: {src.bounds}")
```
